```python
import networkx as nx
import igraph as ig
import numpy as np
import time
import ensembleclustering as CE
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
```

```python
def load_abcd_graph(size, data_dir=""):
    """
    Reads the ABCD graph files and constructs an igraph graph.
    Args:
        size (int): Graph size identifier (e.g., 100, 200, etc.).
        data_dir (str): Directory where ABCD graph files are stored.

    Returns:
        g (igraph.Graph): The loaded graph.
        gt_labels (list): List of ground truth community labels.
    """
    degreefile = os.path.join(data_dir, f"deg_{size}.dat")
    communitysizesfile = os.path.join(data_dir, f"cs_{size}.dat")
    communityfile = os.path.join(data_dir, f"com_{size}.dat")
    networkfile = os.path.join(data_dir, f"edge_{size}.dat")

    # Load the graph edges
    g = ig.Graph.Read_Ncol(networkfile, directed=False)

    # Load community labels (1-based indexing, shift to 0-based)
    c = np.loadtxt(communityfile, dtype='uint16', usecols=(1))
    c -= 1  # Convert to zero-based

    # Assign communities to nodes
    g.vs['comm'] = [c[int(v['name']) - 1] for v in g.vs]

    # Store ground truth communities
    gt_labels = {int(v.index): g.vs[v.index]['comm'] for v in g.vs}

    return g, gt_labels
```

```python
# List of node sizes to test
node_sizes = [100, 200, 400, 800, 1600]
time_results = {algo: [] for algo in ["ECG", "Leiden", "Louvain", "Infomap", "Label Propagation", "Girvan-Newman", "CNM"]}
```

```python
def measure_time(algorithm, graph, *args, **kwargs):
    start_time = time.time()
    if graph is not None:
        result = algorithm(graph, *args, **kwargs)
    else:
        result = algorithm(*args, **kwargs)
    end_time = time.time()
    return result, end_time - start_time
```

```python
def compute_clustering_times(graph, gt_labels):
    # --- 1) Louvain ---
    _, time_louvain = measure_time(ig.Graph.community_multilevel, graph)
    time_results["Louvain"].append(time_louvain)
    
    # --- 2) Leiden ---
    _, time_leiden = measure_time(ig.Graph.community_leiden, graph, objective_function='modularity')
    time_results["Leiden"].append(time_leiden)
    
    # --- 3) Infomap ---
    _, time_infomap = measure_time(ig.Graph.community_infomap, graph)
    time_results["Infomap"].append(time_infomap)
    
    # --- 4) Label Propagation ---
    _, time_lp = measure_time(ig.Graph.community_label_propagation, graph)
    time_results["Label Propagation"].append(time_lp)
    
    # --- 5) Girvan-Newman ---
    gn_hierarchy, time_gn = measure_time(ig.Graph.community_edge_betweenness, graph)
    time_results["Girvan-Newman"].append(time_gn)
    
    # --- 6) CNM ---
    cnm_hierarchy, time_cnm = measure_time(ig.Graph.community_fastgreedy, graph)
    time_results["CNM"].append(time_cnm)
    
    # --- 7) ECG using ClusterEnsembles ---
    labels_louvain = graph.community_multilevel().membership
    labels_infomap = graph.community_infomap().membership
    labels_lp = graph.community_label_propagation().membership
    base_clusterings = np.array([labels_louvain, labels_infomap, labels_lp])
    
    _, time_ecg = measure_time(CE.cluster_ensembles, None, base_clusterings, verbose=False, nclass=5)
    time_results["ECG"].append(time_ecg)
```

```python
# Process each ABCD-like graph size
for n in node_sizes:
    g, gt_labels = load_abcd_graph(n)
    compute_clustering_times(g, gt_labels)
```

```python
# Plot time complexity results
plt.figure(figsize=(10, 6))
for algo, times in time_results.items():
    plt.plot(node_sizes, times, marker='o', label=algo)
plt.xlabel("Number of Nodes")
plt.ylabel("Execution Time (seconds)")
plt.title("Clustering Algorithm Execution Time Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
